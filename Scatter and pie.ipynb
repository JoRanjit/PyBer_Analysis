```python
import matplotlib.pyplot as plt
import numpy as np
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

#list comprehension

#ysize = []
#for y in y_axis:
#    ysize.append(y*3)

ysize = [ysize*3 for ysize in y_axis]
```

```python
plt.scatter(y_axis,x_axis,c='red',label = 'Chicago', s = ysize, edgecolor = 'black')
plt.ylabel("Months")
plt.xlabel("Avg Fare ($)")
plt.title("Average PyBer Fare")
plt.xlim(0,50)
plt.gca().invert_yaxis()
plt.legend()
plt.show()
```

```python
new_ysize = [new_ysize*5 for new_ysize in y_axis]
fig, ax = plt.subplots()
ax.scatter(y_axis,x_axis, s= new_ysize,c= 'skyblue', edgecolor = 'black',linewidth = 2, label = 'Boston')
ax.set_title("Average Pyber Fare")
ax.set_xlabel("Avg Fare ($)")
ax.set_ylabel("Date")
ax.invert_yaxis()
ax.set_xlim(0,50)
ax.legend()
```

```python
explode = (0,0,0.2,0,0,0,0,0,0,0,0,0)
plt.subplots(figsize = (8,8))
plt.pie(y_axis, labels = x_axis, explode = explode, autopct = "%1.1f%%")
ax.invert_yaxis()
plt.show()
```

```python
# this part picks the 2 largest numbers and then creates the explode list 
# this will in turn explode the 1st and 2nd highest in the list
 
new_yaxis=[]
for y in y_axis:
    if y != max(y_axis):
        new_yaxis.append(y)
new_yaxis

#added all items to new_yaxis except the Max
#then assigned the new max as the max of the new_yaxis

new_max = max(new_yaxis) 

explode=[]
for i in y_axis:
    print(i)
    if i >= new_max:
        explode.append(0.1)
    else:
        explode.append(0)
explode
```

```python
fig,ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis, labels = x_axis,autopct = "%1.1f%%",startangle = 90,explode=explode, shadow = True)
ax.invert_xaxis()
plt.show()
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis,y_axis,yerr=stdev, capsize=3)
```
